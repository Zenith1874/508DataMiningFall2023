```python
#5.Calculate and print how many people have “default payment” = 1 and how many peoplehave “default payment” = 0.
df = df.rename(columns={'default payment next month': 'default_payment'})

target_value0 = 1
target_value1 = 0
count_equal_to_target1 = (df['default_payment'] == target_value1).sum()
#Calculate and print how many people have “default payment” = 1 and how many people have “default payment” = 0.
grouped_df = df.groupby('default_payment')['ID'].count()       ## group
print(grouped_df)

count_equal_to_target0 = (df['default_payment'] == target_value0).sum()
print("Number of “default payment” = 1 equal to {}: {}".format(target_value0, count_equal_to_target1))
print("Number of “default payment” = 0 equal to {}: {}".format(target_value1, count_equal_to_target0))
```

```python
#6. Calculate and print how many people who are married and have “default payment” = 1
condition1 = df['MARRIAGE']  == 1
condition2 = df['default_payment']  == 1
count_with_conditions = (condition1 & condition2).sum()
print(" people who are married and have default payment = 1: {}".format(count_with_conditions))
```

```python
#7. Calculate and print how many people whose age is greater than 30 and have “default payment” = 1

condition1 = df['AGE']  > 30
condition2 = df['default_payment']  == 1
count_with_conditions = (condition1 & condition2).sum()
print(" people whose age is greater than 30 and have “default payment” = 1: {}".format(count_with_conditions))
```

```python
#8. Calculate the average value of the “LIMIT_BAL” column when gender is male, and when gender is female.

average_debt_male = df[df['SEX'] == 1]['LIMIT_BAL'].mean()
average_debt_female = df[df['SEX'] == 2]['LIMIT_BAL'].mean()
print("Average LIMIT_BAL for male:", average_debt_male)
print("Average LIMIT_BAL for female:", average_debt_female)
```
