```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Age': [25, 30, 22, 35, 28, 40, 29],
        'Late_Payment': [100, 200, 150, 300, 120, 180, 90],
        'Default_Payment': [50, 80, 70, 120, 60, 100, 40]}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
df[df['Age'] <= 30]['Late_Payment'].plot(kind='hist', bins=10, color='blue', alpha=0.7, label='Late Payment')
df[df['Age'] > 30]['Default_Payment'].plot(kind='hist', bins=10, color='orange', alpha=0.7, label='Default Payment')

plt.xlabel('Payment Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Payment for Different Age Groups')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Age': [25, 30, 22, 35, 28, 40, 29],
        'Late_Payment': [100, 200, 150, 300, 120, 180, 90],
        'Default_Payment': [50, 80, 70, 120, 60, 100, 40]}
df = pd.DataFrame(data)

# 30
df_age_less_than_30 = df[df['Age'] <= 30]

# 30
df_age_greater_than_30 = df[df['Age'] > 30]

plt.figure(figsize=(10, 6))

# 30Default_Payment
plt.hist(df_age_less_than_30['Default_Payment'], bins=10, color='blue', alpha=0.7, label='Age <= 30')

# 30Default_Payment
plt.hist(df_age_greater_than_30['Default_Payment'], bins=10, color='orange', alpha=0.7, label='Age > 30')

plt.xlabel('Default Payment')
plt.ylabel('Frequency')
plt.title('Histogram of Default Payment for Different Age Groups')
plt.legend()
plt.show()
```
